```python
import pandas as pd

# Load CSV files and add 'channel' column
facebook_df = pd.read_csv('Facebook.csv')
facebook_df['channel'] = 'Facebook'

tiktok_df = pd.read_csv('TikTok.csv')
tiktok_df['channel'] = 'TikTok'

google_df = pd.read_csv('Google.csv')
google_df['channel'] = 'Google'

# Merge dataframes
merged_df = pd.concat([facebook_df, tiktok_df, google_df], ignore_index=True)

# Save to new CSV
merged_df.to_csv('merged_social_media.csv', index=False)
merged_df.head()
```

```python
business_df = pd.read_csv('business.csv')
business_df.head()
```

```python
# Calculate sums
total_revenue = merged_df['attributed revenue'].sum()
total_spend = merged_df['spend'].sum()

# Print results
print("Sum of attributed revenue:", total_revenue)
print("Sum of spend:", total_spend)
print("Sum(attributed revenue)/Sum(spend):", total_revenue / total_spend)

# Calculate sum of new customers from business_df
total_customers = business_df['new customers'].sum()
print("Sum(new customers):", total_customers)
print("Sum(spend)/Sum(new customers):", total_spend / total_customers)
```
